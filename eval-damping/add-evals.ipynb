```python
import pandas as pd
import chess
import chess.engine
from tqdm import tqdm

tqdm.pandas()
```

```python
data = pd.read_csv('data/damping.csv')
data.head()
```

```python
engine = chess.engine.SimpleEngine.popen_uci(
    "C:\\Users\\maxen\\Repos\\Stockfish\\stockfish_x86-64-avx2.exe"
)


def deep_eval(position: str):
    board = chess.Board(position)
    info = engine.analyse(board, chess.engine.Limit(depth=4))
    return info["score"].white().score(mate_score=32000)

data["deep_eval"] = data["position"].progress_apply(deep_eval)
engine.quit()
```

```python
data.to_csv("data/damping_with_evals.csv")
data.head()
```
